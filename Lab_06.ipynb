### **X-ray Image Processing with Edge Enhancement**

This code processes an X-ray image to improve its clarity. First, the image is loaded in grayscale to keep things simple.

Next, a Gaussian Blur is applied to reduce noise, making the image smoother and easier to analyze. This helps remove unwanted details while preserving important structures.

Then, a Laplacian filter is used to detect and enhance edges. This makes the boundaries of bones and tissues more visible, helping doctors see critical details.

Finally, the original, denoised, and edge-enhanced images are displayed side by side using Matplotlib. This allows for a clear comparison of how the enhancements improve the image.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the X-ray image in grayscale
xray_image = cv2.imread("/content/Scanerio 01.png", cv2.IMREAD_GRAYSCALE)
# Apply a different Gaussian Blur kernel size for denoising
blurred_xray = cv2.GaussianBlur(xray_image, (7, 7), 1.5)
# Use Laplacian filter for edge enhancement with a modified kernel
edges_xray = cv2.Laplacian(blurred_xray, cv2.CV_64F, ksize=3)
# Convert to absolute values for proper visualization
enhanced_xray = cv2.convertScaleAbs(edges_xray)
# Display results using Matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(xray_image, cmap='gray')
plt.title("Original X-ray")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(blurred_xray, cmap='gray')
plt.title("Denoised X-ray (Gaussian Blur)")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(enhanced_xray, cmap='gray')
plt.title("Enhanced Edges (Laplacian)")
plt.axis("off")
plt.tight_layout()
plt.show()
```

### **Road Lane Detection Using Canny Edge Detection**

This code enhances a road image to detect lane markings more clearly. First, it loads the image in grayscale and applies a Gaussian Blur to remove noise and smooth the surface. Then, it uses the Canny edge detection method with adjusted thresholds to highlight lane edges. Finally, the original, smoothed, and edge-detected images are displayed side by side for comparison.

```python
import cv2
import matplotlib.pyplot as plt
# Load the grayscale image of the road
road_img = cv2.imread("/content/Scanerio 02.png", cv2.IMREAD_GRAYSCALE)
# Use a different Gaussian kernel for denoising
smoothed_road = cv2.GaussianBlur(road_img, (7, 7), 1)
# Adjust Canny edge detection thresholds
lane_edges = cv2.Canny(smoothed_road, 40, 130)
# Display the processed images using Matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(road_img, cmap='gray')
plt.title("Original Road")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(smoothed_road, cmap='gray')
plt.title("Smoothed Road (Gaussian Blur)")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(lane_edges, cmap='gray')
plt.title("Lane Detection (Canny Edges)")
plt.axis("off")
plt.tight_layout()
plt.show()
```

### **Old Photo Restoration and Enhancement**

This code restores an old photo by reducing noise and enhancing details. First, it loads the image in grayscale and applies a Bilateral filter, which smooths the image while preserving edges. Then, a sharpening filter is applied to enhance details and improve clarity. Finally, the original, filtered, and sharpened images are displayed side by side for comparison.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the grayscale image of an old photo
old_image = cv2.imread("/content/Scanerio 03.png", cv2.IMREAD_GRAYSCALE)
# Use a different Bilateral filter configuration
filtered_image = cv2.bilateralFilter(old_image, 11, 90, 90)
# Define a sharpening kernel with adjusted values
sharpening_filter = np.array([[0, -1, 0],
 [-1, 5, -1],
 [0, -1, 0]])
# Apply sharpening to the filtered image
sharpened_photo = cv2.filter2D(filtered_image, -1, sharpening_filter)
# Display the processed images using Matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(old_image, cmap='gray')
plt.title("Original Photo")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title("Filtered Photo (Bilateral)")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(sharpened_photo, cmap='gray')
plt.title("Enhanced Photo (Sharpened)")
plt.axis("off")
plt.tight_layout()
plt.show()
```

### **Fingerprint Enhancement Using Noise Reduction**

This code enhances a fingerprint image for better clarity and ridge detection. First, it loads the image in grayscale and applies a Median filter to remove noise while preserving details. Then, a Bilateral filter is used to further smooth the image while keeping ridge edges sharp. Finally, the original, denoised, and enhanced images are displayed side by side for comparison.

```python
import cv2
import matplotlib.pyplot as plt
# Load fingerprint image in grayscale
fingerprint = cv2.imread("/content/Scanerio 04.png", cv2.IMREAD_GRAYSCALE)
# Use a different median filter kernel size
denoised_fingerprint = cv2.medianBlur(fingerprint, 7)
# Apply Bilateral filter with modified parameters
enhanced_fingerprint = cv2.bilateralFilter(denoised_fingerprint, 13,
85, 85)
# Display results using Matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(fingerprint, cmap='gray')
plt.title("Original Fingerprint")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(denoised_fingerprint, cmap='gray')
plt.title("Denoised (Median Filter)")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(enhanced_fingerprint, cmap='gray')
plt.title("Enhanced Ridges (Bilateral)")
plt.axis("off")
plt.tight_layout()
plt.show()
```
